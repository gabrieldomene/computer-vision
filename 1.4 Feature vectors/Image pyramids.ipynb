# Image pyramids

To construct the layers of "pyramid", use function `cv2.pyrDown(image)`

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

img = cv2.imread("images/rainbow_flag.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(img)
```

```python
lvl_1 = cv2.pyrDown(img)
lvl_2 = cv2.pyrDown(lvl_1)
lvl_3 = cv2.pyrDown(lvl_2)

fig, axs = plt.subplots(1, 4, figsize=(20, 10))
(ax1, ax2, ax3, ax4) = axs
ax1.set_title("Original")
ax1.imshow(img)

for idx, ax in enumerate(axs, 1):
    ax.set_xlim([0, img.shape[1]])
    ax.set_ylim([0, img.shape[0]])

ax2.set_title("Level 1")
ax2.imshow(lvl_1)

ax3.set_title("Level 2")
ax3.imshow(lvl_2)

ax4.set_title("Level 3")
ax4.imshow(lvl_3);





```
