## HSV Color space, balloons
### Import resources and display

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
%matplotlib inline

img = cv2.imread("images/water_balloons.jpg")
img_cp = np.copy(img)

img = cv2.cvtColor(img_cp, cv2.COLOR_BGR2RGB)
plt.imshow(img);
```

### Plot channels

```python
r, g, b = img[:, :, 0], img[:, :, 1], img[:, :, 2]
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
h, s, v = hsv[:, :, 0], hsv[:, :, 1], hsv[:, :, 2]

f, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(20, 10))

ax1.set_title("R")
ax1.imshow(r, cmap="gray")
ax2.set_title("G")
ax2.imshow(r, cmap="gray")
ax3.set_title("B")
ax3.imshow(r, cmap="gray")

ax4.set_title("H")
ax4.imshow(h, cmap="gray")
ax5.set_title("S")
ax5.imshow(s, cmap="gray")
ax6.set_title("V")
ax6.imshow(v, cmap="gray")

```

### Define pink and hues thresholds

```python
# HSV
lower_hue = np.array([160, 0, 0])
upper_hue = np.array([180, 255, 255])

# RGB
lower_pink = np.array([180, 0, 130])
upper_pink = np.array([255, 255, 230])
```

## Mask the image

```python
# Mask in the hsv space

mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)

masked = np.copy(img)
masked[mask_hsv == 0] = [0, 0, 0]

plt.imshow(masked);
```

```python
# Mask in the rgb space

mask_rgb = cv2.inRange(img, lower_pink, upper_pink)

masked = np.copy(img)
masked[mask_rgb == 0] = [0, 0, 0]

plt.imshow(masked);
```
