## Day and night image classifier
The dataset consists of 200 RGB color images in two categories: day and night. The goal is to build a classifier that can label these images as day or night

### Import resources

```python
import cv2
import helpers

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

## Training and testing data

```python
img_dir_training = "day_night_images/training/"
img_dir_test = "day_night_images/test/"
```

## Load the datasets


```python
img_list = helpers.load_dataset(img_dir_training)
```

## 1. Visualize the input images

```python
img_index = 0
selected_img = img_list[img_index][0]
selected_label = img_list[img_index][1]

print("Shape {} - Label {}" .format(selected_img.shape, selected_label))
```

```python
import random
random.shuffle(img_list)

for idx, img in enumerate(img_list):
    if img_list[idx][1] == "day":
        day_img = img_list[idx][0]
        break
        
for idx, img in enumerate(img_list):
    if img_list[idx][1] == "night":
        night_img = img_list[idx][0]
        break

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 20))

ax1.set_title("Day")
ax1.imshow(day_img)
ax2.set_title("Night")
ax2.imshow(night_img);
```
