### 2.Pre-process the data
After loading each image, you need to standardize the input and output

#### Standardize the input

```python
'''This function should take in and RGB image and return a new standardized version of 600x1100
'''

def standardize_input(image):
    return cv2.resize(image, (1100, 600), interpolation=cv2.INTER_CUBIC)
```

#### Standardize the output
Return binary numerical value for the labels

```python
def standardize_label(label):
    return 1 if label == "day" else 0
```

#### Construct a standardized_list of input images and output labels.
This function takes in a list of img-label pairs and outputs a standardized list of resized images and numerical labels.


```python
def standardized_list(img_list):
    standard_list = []
    for idx, img_tuple in enumerate(img_list):
        img, label = img_tuple
        
        resized_img = standardize_input(img)
        binary_label = standardize_label(label)
        standard_list.append((resized_img, binary_label))
    return standard_list
```
