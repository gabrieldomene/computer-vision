## Color threshold green screen

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np
import cv2

%matplotlib inline
```

### Read and display

```python
img = mpimg.imread("images/car_green_screen.jpg")

print("Shape {}" .format(img.shape))
```

```python
plt.imshow(img);
```

### Define color threshold

```python
lower_green = np.array([0, 220, 0])
upper_green = np.array([100, 255, 100])
```

### Define mask

```python
mask = cv2.inRange(img, lower_green, upper_green)


masked = np.copy(img)
masked[mask != 0] = [0, 0, 0]

plt.imshow(masked, cmap="gray");
```

### Mask and bg img

```python
bg = mpimg.imread("images/sky.jpg")
background = bg.copy()
w, h= mask.shape

cropped_bg = background[0:w, 0:h]
cropped_bg[mask == 0] = [0, 0, 0]

final = cropped_bg + masked
plt.imshow(final);
```
