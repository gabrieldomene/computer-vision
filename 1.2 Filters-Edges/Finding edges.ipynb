## Edge detection

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
%matplotlib inline

img = mpimg.imread("images/curved_lane.jpg")

plt.imshow(img)
```

## Convertion

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray");
```

## Custom kernel

Challanges:
    * Create a filter for blur the image
    * Create sobel x
    * Create sobel y

```python
def blur_img(img):
    kernel = np.array([[1, 4, 6, 4, 1],
                       [4, 16, 24, 16, 4],
                       [6, 24, 36, 24, 6],
                       [4, 16, 24, 16, 4],
                       [1, 4, 6, 4, 1]])/256
    blur = cv2.filter2D(img, -1, kernel)

    return blur

def sobel_x(img):
    kernel = np.array([[1, 0, -1],
                       [2, 0, -2],
                       [1, 0, -1]])
    edges_x = cv2.filter2D(img, -1, kernel)
    
    return edges_x

def sobel_y(img):
    kernel = np.array([[1, 2, 1],
                       [0, 0, 0],
                       [-1, -2, -1]])
    edges_y = cv2.filter2D(img, -1, kernel)
    
    return edges_y

blurred = blur_img(gray)
edges_x = sobel_x(gray)
edges_y = sobel_y(gray)
edges_x_blur = sobel_x(blurred)
edges_y_blur = sobel_y(blurred)

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(15, 15))

ax1.set_title("Original")
ax1.imshow(gray, cmap="gray")
ax2.set_title("Blurred")
ax2.imshow(blurred, cmap="gray")

ax3.set_title("Sobel X")
ax3.imshow(edges_x, cmap="gray")
ax4.set_title("Sobel Y")
ax4.imshow(edges_y, cmap="gray")

ax5.set_title("Blur Edges X")
ax5.imshow(edges_x_blur, cmap="gray")
ax6.set_title("Blur Edges Y")
ax6.imshow(edges_y_blur, cmap="gray")

```
