## Creating a filter edge detection
### Imports and display

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

img = cv2.imread("images/city_hall.jpg")
img_cp = np.copy(img)

img_cp = cv2.cvtColor(img_cp, cv2.COLOR_BGR2RGB)
plt.imshow(img_cp)
```

```python
# Grayscale

gray = cv2.cvtColor(img_cp, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
```

## Custom kernel

```python
# kernel 3x3
# Sobel filter
sobel_x = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

filtered = cv2.filter2D(gray, -1, sobel_x)
plt.imshow(filtered, cmap="gray");
```

## Create binary image

```python
# Create a threshold that sets all the filtered pixels to white

_, binary_image = cv2.threshold(filtered, 120, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap="gray");
```
