```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
img = cv2.imread("images/brain_MR.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
```

## Canny edges detection

```python
# Wide and tight values for threshold lines (low and upper)

wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 200, 240)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title("Wide values")
ax1.imshow(wide, cmap="gray")

ax2.set_title("Tight values")
ax2.imshow(tight, cmap="gray");
```

```python
flower = cv2.imread("images/sunflower.jpg")
flower = cv2.cvtColor(flower, cv2.COLOR_BGR2RGB)

gray_flower = cv2.cvtColor(flower, cv2.COLOR_RGB2GRAY)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.imshow(flower)
ax2.imshow(gray_flower, cmap="gray");
```

```python
from random import randint

wide = cv2.Canny(gray_flower, randint(1, 50), randint(75, 150))
tight = cv2.Canny(gray_flower, randint(175, 200), randint(210, 245))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title("Wide canny")
ax1.imshow(wide, cmap="gray")

ax2.set_title("Tight canny")
ax2.imshow(tight, cmap="gray");
```
