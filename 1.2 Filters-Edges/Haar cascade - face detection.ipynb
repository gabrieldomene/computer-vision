## Face detection using OpenCV
Old but popular method

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread("images/multi_faces.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20, 10))
plt.imshow(img)
```

Since face detection don't need much features based on color channel, convert it to gray

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(15, 10))
plt.imshow(gray, cmap="gray");
```

## Load the data
`haarcascade_frontalface_default.xml` is a fully trained architecture of face detector.

* Note on parameters
The number of detected face is determined by the function `detectMultiScale`, the function inputs are (image, scaleFactor, minNeighbors) and smaller scaleFactor often detects more faces and lower value for minNeighbors. Modify these values if the image does not show good results.

The output is an array of detections (bounding boxes of face)

```python
face_cascade = cv2.CascadeClassifier("detector_architectures/haarcascade_frontalface_default.xml")

faces = face_cascade.detectMultiScale(gray, 4, 6)

print("Detected {} faces in the image" .format(len(faces)))
```

```python
img_detections = np.copy(img)

for(x, y, w, h) in faces:
    cv2.rectangle(img_detections, (x, y), (x+w, y+h), (255, 0, 0), 4)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title("Input")
ax1.imshow(img)

ax2.set_title("Detections")
ax2.imshow(img_detections);
```
